```python
import pandas   as pd
import numpy    as np   
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_excel('train.csv.xlsx')
display(df.head())
```

```python
df.isnull().sum()   
```

```python
#bar plot showing survival rate by passenger class

sns.barplot(x='Pclass', y='Survived', data=df, palette='Blues')

```

```python
# scatter plot of fare vs Age colored by survival sates
sns.scatterplot(y='Fare', x='Age', hue='Survived', data=df, palette='coolwarm')
 
```

```python
#passenger counts by embarked port as pie chart
df['Embarked'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, )
plt.title('Passenger Counts by Embarked Port')
```

```python
#survival rates by family size(SIBSP + Parch)
df['FamilySize'] = df['SibSp'] + df['Parch']
sns.barplot(x='FamilySize', y='Survived', data=df, palette='pastel')
```

```python

# heatmaps for survival rates by class and sex

heatmap_data = df.pivot_table(index='Pclass', columns='Sex', values='Survived')
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm')
plt.title('Survival Rates by Class and Sex')




```

```python
#fare disteribution by survival status



sns.boxplot(x='Pclass', y='Fare', data=df, palette='pastel')
```
