sine and cosine


```python
x = np.arange(0,6*np.pi, 0.01)
y_sin = np.sin(x)
y_cos = np.cos(x)


plt.subplot(2, 2, 1)
plt.plot(x, y_cos, label='Cosine Wave')
plt.title('Cosine Wave')
plt.subplot(2, 2, 4)
plt.plot(x, y_sin, color ="red",label='Sine Wave')
plt.title('Sine Wave');
```

```python
x = [1,2,3,4]
y =[5,7,4,6]
x2 = [1,2,3,4]
y2 = [10,14,12,13]
plt.plot(x, y, label='Line 1', color='blue')
plt.plot(x2, y2, label='Line 2', color='red')
plt.legend();
```

bar chart


```python
x = [1,2,3,4]
y =[5,7,4,6]
x2 = [1,2,3,4]
y2 = [10,14,12,13]
plt.subplot(2,1, 1)
plt.bar(x, y, label='Line 1', color='blue')
plt.subplot(2,1, 2)
plt.bar(x2, y2, label='Line 2', color='red')
plt.legend();
```
