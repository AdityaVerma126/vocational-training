```python
import pandas   as pd
import numpy    as np   
import matplotlib.pyplot as plt
import seaborn as sns   
```

opn the datset


```python
df = pd.read_csv('heart.csv')
display(df.head())
```

calulate mean median mode

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
#calculate mean 
print("\nMean of each column:")
print(df.mean())
plt.boxplot(df["age"]);
```

```python
# Calculate median
print("\nMedian of each column:")
print(df.median());
```

```python
#calculate the mode
print("\nMode of each column:")
print(df.mode().iloc[0])
```

```python
plt.hist(df['trestbps'], bins=10)
```

```python
#calculate standard deviation
print("\nStandard deviation of each column:")
print(df.std())
plt.hist(df.std(), bins=10, color='blue', alpha=0.7)
```

```python
plt.hist(df['age'], bins=20, color='blue', edgecolor='black')
```

create heatmaps to find coreelations

```python
#CORRELATIONS
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.1f')
```

scatter plot between age and max heart rate

```python
plt.scatter(df['age'], df['thal'], alpha=0.5 )
```

age and chol

```python
plt.scatter(df['age'], df['chol'], alpha=0.5, color='red')
```

patients with heart disease using pie chart

```python
plt.pie(df['target'].value_counts(), labels=['No Heart Disease', 'Heart Disease'], autopct='%1.1f%%', startangle=140) ;
```

grouping data by chest pain tyoe and analyse the avg cholestrol and heart rate values

```python
df.groupby('cp')[[ 'chol', 'thalach']].mean().plot(kind='bar',)
```

```python
#AVG REATING BLOOD OPRESSURE FOR NMALES AND FEMALES
group = df.groupby('sex')
plt.bar(group['trestbps'].mean(), group['trestbps'].mean(), color=['red', 'orange'])
```
